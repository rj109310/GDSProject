```python
import pandas as pd
df = pd.read_csv("./FPData2.csv")
df.head()
df
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df.columns = ['Country', 'year', 'ladder', 'gdp', 'support', 'life_expec', 'freedom', 'generosity','corruption','positive', 'negative']
df.head()
```

```python
y = df.groupby('Country').count()
less = y[y['year'] <= 10].index.tolist()
df_update = df[~df['Country'].isin(less)]
```

```python
df_update.groupby('Country').count()
```

```python
df_update.groupby('year').size()
```

```python
df_means = df_update.groupby(['year']).mean()
df_means
```

```python
df_means1 = df_means.drop([2005])
df_means2 = df_means1.drop([2006])
df_means3 = df_means2.drop([2007])
df_means4 = df_means3.drop([2008])
df_means5 = df_means4.drop([2009])
df_means6 = df_means5.drop([2010])
df_means7 = df_means6.drop([2020])
df_means8 = df_means7.drop([2021])
df_means9 = df_means8.drop([2022])
df_means9
```

```python
x_labels = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
#Plotting the means for each year, removing 2005, due to missing data
fig, axs = plt.subplots(2, 4, figsize=(14,8))
axs[0, 0].plot(x_labels, df_means9['ladder'])
axs[0, 0].set_title('Ladder')
axs[0, 1].plot(x_labels, df_means9['generosity'])
axs[0, 1].set_title('Generosity')
axs[1, 0].plot(x_labels, df_means9['life_expec'])
axs[1, 0].set_title('Life Expectency')
axs[1, 1].plot(x_labels, df_means9['freedom'])
axs[1, 1].set_title('Freedom')
axs[0, 2].plot(x_labels, df_means9['support'])
axs[0, 2].set_title('Social Support')
axs[0, 3].plot(x_labels, df_means9['corruption'])
axs[0, 3].set_title('Corruption')
axs[1, 2].plot(x_labels, df_means9['positive'])
axs[1, 2].set_title('Positive Affect')
axs[1, 3].plot(x_labels, df_means9['negative'])
axs[1, 3].set_title('Negative Affect')
```
